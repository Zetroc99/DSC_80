### Missingness, another look
.

#### The missing values in a column can have 3 mechanisms for such missingness:

#### 1) An external reason, completely unrelated to the survey/table/data generating process. Then, it's MCAR. Usually rare. 
.
.
#### 2) Other column(s) in the table. 
.
#### a) If one can explain each and every missingess based on other columns, then it's MD. 
.
#### b) If there's some probabilistic dependency on other columns, then it's MAR.
.
.
#### 3) The column itself. Then it's NMAR / NI.

### Imputation:
.
#### Mean (or single-value) imputation: Impute with the mean of the column. Preserves mean, reduces variance. Biases mean usually, since MCAR is rare.
.
#### Probabilistic imputation: Impute with values depending on the probability of their occurrence. If MAR on another column, draw from the empirical distribution of each group of that column!

```python
rng = np.random.RandomState(10)  # deterministic random data

a = np.hstack((rng.normal(size=1000),

               rng.normal(loc=5, scale=2, size=1000)))
np.histogram(a)
```

```python
probs = np.histogram(a)[0] / sum(np.histogram(a)[0])
bins = np.histogram(a)[1]
```

```python
probs
```

```python
bin_choice = np.random.choice(range(len(probs)), p = probs)
bin_choice
```

```python
np.random.uniform(bins[bin_choice], bins[bin_choice+1])
```
